### Визуализации распределения параметров
 функция  hist_with_stat_lines для постороения гистограммы в рамкахах  нормальности, т.е. в рамках q1 - 1.5iqr : q3 + 1.5iq
 c нанесением линий q1, q3, среднего и медианы. Это решение не лишено критики, т.к распределения далеко не всегда нормальные и симметричные. Но в этой учебной работе всеже применим квартильные методы для анализа фильтрации данных.

```python
def hist_with_stat_lines(df, df_col, dtitle, dxlabel, dylabel, bins = 0, kde=False):
    df_col_description =  df_col.describe()
    mean = df_col_description[1]
    std = df_col_description[2]
    df_min = df_col_description[3]
    q1 = df_col_description[4]
    median = df_col_description[5]
    q3 = df_col_description[6]
    df_max = df_col_description[7]
    iqr = q3-q1
    left_border = q1 - iqr*1.5
    # усы обрезаются минимальным и максимальным значением параметра сооветственно
    if left_border < df_min:
        left_border = df_min
    right_border = q3 + iqr*1.5
    if right_border > df_max:
        right_border = df_max
        
    if bins == 0:
        bins = right_border - left_border
        if bins>100:
            bins = 100
        bins = int(bins)
    else:
        bins = bins

    fig = sns.histplot(df, x=df_col, bins=bins, kde=kde  ) # binwidth = bins/10 #kde=True  edgecolor='g' bins=bins  binwidth = bins/50
    fig.set(xlim=(left_border, right_border)) 
    plt.title(dtitle)
    plt.xlabel(dxlabel)
    plt.ylabel(dylabel)
    plt.axvline(x=mean,linewidth=1, color='r', linestyle='-' , label= 'mean_value')
    plt.axvline(x=median,linewidth=1, color='b', linestyle='-', label= 'median_value')
    plt.axvline(x=q1,linewidth=1, color='g', linestyle='--',label= 'Q1_value')
    plt.axvline(x=q3,linewidth=1, color='g', linestyle='--',label= 'Q3_value')
    plt.legend(loc = 'upper right')
    return 
```
