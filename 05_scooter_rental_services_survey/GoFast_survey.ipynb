Вы аналитик популярного сервиса аренды самокатов GoFast. Вам передали данные о некоторых пользователях из нескольких городов, а также об их поездках. Проанализируйте данные и проверьте некоторые гипотезы, которые могут помочь бизнесу вырасти.
Чтобы совершать поездки по городу, пользователи сервиса GoFast пользуются мобильным приложением. Сервисом можно пользоваться:
-	без подписки 
--	абонентская плата отсутствует;
--	стоимость одной минуты поездки — 8 рублей;
--	стоимость старта (начала поездки) — 50 рублей;
-	с подпиской Ultra 
--	абонентская плата — 199 рублей в месяц;
--	стоимость одной минуты поездки — 6 рублей;
--	стоимость старта — бесплатно.


Описание данных:
В основных данных есть информация о пользователях, их поездках и подписках.

**Пользователи — users_go.csv**

- `user_id` - уникальный идентификатор пользователя
- `name`- имя пользователя
- `age`- возраст
- `city`- город
- `subscription_type`- тип подписки (free, ultra)

**Поездки — rides_go.csv**
- `user_id` - уникальный идентификатор пользователя
- `distance` - расстояние, которое пользователь проехал в текущей сессии (в метрах)
- `duration` - продолжительность сессии (в минутах) — время с того момента, как пользователь нажал кнопку «Начать поездку» до момента, как он нажал кнопку «Завершить поездку»
- `date` - дата совершения поездки

**Подписки — subscriptions_go.csv**

- `subscription_type` - тип подписки
- `minute_price`- стоимость одной минуты поездки по данной подписке
- `start_ride_price`- стоимость начала поездки
- `city`- город
- `subscription_fee`- стоимость ежемесячного платежа

## Проверка гипотез

### Тратят ли пользователи с подпиской больше времени на поездки
Сформулируем H0 гипотезу средние выборок равны. И соответственно H1:среднее выборки пользователей с подпиской больше среднего выборки, где пользователи не покупали подписку. 

```python
#df_free_users.duration
#df_ultra_users.duration
alpha = 0.05 # уровень статистической значимости

results = st.ttest_ind( df_ultra_users.duration, df_free_users.duration, alternative='greater')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: пользователи с подпиской тратят больше времени на поездки ')
else:
    print('Не получилось отвергнуть нулевую гипотезу: исследуемой разница не подтверждена')
display(df_ultra_users.duration.mean())
df_free_users.duration.mean()
```

<div class="alert alert-success">
<b>Комментарий ревьюера:</b>☀️
    
Отлично, что формулируешь нулевую, альтернативную гипотезу, это важно, выводы верны, ещё могу порекомендовать дополнительно проверять себя и выводить среднее проверяемых переменных или распределения

Да, по результатам T-теста, даже при уровне статистической значимости 0.01, пользователи с подпиской тратят больше времени на поездки.

### Не превышает ли среднее расстояние, которое проезжают пользователи с подпиской за одну поездку  3130 метров
Сформулируем H0 - выборочное среднее равно 3130. Или H1  выборочное среднее больше 3130

```python
h0_value = 3130
alpha = 0.05# уровень статистической значимости
results =st.ttest_1samp( df_ultra_users.distance, h0_value, alternative='greater')
print('p-значение:', results.pvalue) 
if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу: среднее расстояние больше 3130м')
else:
    print('Не получилось отвергнуть нулевую гипотезу: среднее расстояние не превышает 3130м ')
```

### Выше ли помесячная выручка от пользователей с подпиской чем выручка от пользователей без подписки
Сформулируем H0 - средниене не отличаются, H1 - помесячная выручка от пользователей с подпиской выше, чем выручка от пользователей без подписки. 


```python
# Подготовим серии данных
#ultra_users_sum_receipts_by_mounth = summary_df.query('subscription_fee == 199')\
#                                               .groupby('month_of_date')['sum_receipts'].sum()
#free_users_sum_receipts_by_mounth = summary_df.query('subscription_fee == 0')\
#                                               .groupby('month_of_date')['sum_receipts'].sum()

ultra_users_sum_receipts = summary_df.query('subscription_type == "ultra"')['sum_receipts']
free_users_sum_receipts = summary_df.query('subscription_type == "free"')['sum_receipts']
alpha = 0.05 # уровень статистической значимости

#results = st.ttest_ind( ultra_users_sum_receipts, free_users_sum_receipts, \
#                        equal_var=False, alternative='greater')
results = st.ttest_ind( ultra_users_sum_receipts, free_users_sum_receipts, \
                         alternative='greater')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: помесячная выручка от пользователей с подпиской выше ')
else:
    print('Не получилось отвергнуть нулевую гипотезу:  помесячная выручка от пользователей с подпиской не выше')
display (ultra_users_sum_receipts.sum())
display(free_users_sum_receipts.sum())
display (ultra_users_sum_receipts.count())
display(free_users_sum_receipts.count())
display (ultra_users_sum_receipts.mean())
display(free_users_sum_receipts.mean())
```

помесячная выручка от пользователей с подпиской выше, в то в ремя как суммарно по мясяцам и всему периоду, за который представлены данные доход от пользователей без подписки выше, потому что они совершили примерно на треть больше поездок

Выясните, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно 5 %. Подберите параметры распределения, описывающего эту ситуацию, постройте график распределения и сформулируйте ответ на вопрос о количестве промокодов


```python
from scipy.stats import poisson
from scipy.stats import binom , norm

# зададим функцию, которая выдаёт <num> чисел,
# равномерно распределённых от <start> до <stop>
# (понадобится для визуализации нормального распределения)
def linspace(start, stop, num):
    step = (stop - start) / (num - 1)
    result = []
    for i in range(num):
        result.append(start + step * i)
    return result
 
# задаём параметры биномиального распределения
n = 1165 #тысячи уже достаточно, чтобы убедится что апроксимация хорошо работает.
p = 0.10

# границы промежутка, в котором будем строить аппроксимацию -
# от n*p минус 4 ст.отклонения бином.распределения
# до n*p плюс 4 ст.отклонения бином.распределения
# (там, где вероятности значений почти не равны нулю)
left = int(n * p - 4 * sqrt(n * p * (1 - p)))
right = int(n * p + 4 * sqrt(n * p * (1 - p)))
 
# зададим набор значений биномиальной величины на выбранном промежутке:
# все неотрицательные целые числа от left до right
x = list(range(max(0, left), right))
# рисуем график биномиального распределения
ax = plt.subplot()
# используем синие о-образные маркеры и размер маркера 5
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=5,
	      label='вероятности биномиального распределения')

# зададим мат.ожидание и ст.отклонение нормального распределения равными
# мат.ожиданию и ст.отклонению биномиального распределения
mu = n * p
sigma = sqrt(n * p * (1 - p))
 
# рисуем график нормального распределения:
# получим список из 30 чисел от left до right
x = linspace(left, right, 30)
# используем красную линию ширины 3 и прозрачности 0.3
ax.plot(x, norm.pdf(x, mu, sigma), 'r-', lw=3, alpha=0.3, label='плотность вероятности нормального распределения')
 
# зададим подписи по оси x в пределах ± 3 ст.отклонений от мат.ожидания
# (последний аргумент в методе range - интервал между подписями)
x_ticks = list(range(int(mu - 3 * sigma), int(mu + 3 * sigma), 5))
ax.set_xticks(x_ticks)

# настраиваем оптимальное расположение легенды и выводим график
ax.legend(loc='best')
plt.xticks(rotation=90)
plt.show()

# выводим, выполняется ли наше условие аппроксимации:
# лежит ли mu ± 3 sigma в промежутке от 0 до n:
if (mu - (3 * sigma) >= 0) and ((mu + (3 * sigma) <= n)):
    print('Условие выполняется')
else:
    print('Условие не выполняется')
```
