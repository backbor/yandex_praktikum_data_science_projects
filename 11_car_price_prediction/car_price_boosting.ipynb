```python
#ф-ия расчета  и визуализации процента пропусков.
def missing_data_hist(data):
    part_of_nans = (data.isna().sum() / data.shape[0])*100 
    part_of_nans = part_of_nans.sort_values(ascending = True) 
    ax = plt.gca()
    ax.barh(part_of_nans.index, part_of_nans.values, align='center')
    ax.set_title('Количество пропущенных значений по столбцам в %', size = 12)
    ax.set_xlim(0, 110)
    ax.set_ylabel('Имена столбцов')
    ax.set_xlabel('Шкала в процентах')
    plt.size=(19,10)
    # Place a label for each bar\n",
    rects = ax.patches
    for rect in rects:
        # Get X and Y placement of label from rect
        x_value = rect.get_width()
        y_value = rect.get_y() + rect.get_height() / 2
        # Number of points between bar and label; change to your liking
        space = 5
        # Vertical alignment for positive values
        ha = 'left'
        # If value of bar is negative: place label to the left of the bar
        if x_value < 0:
            # Invert space to place label to the left
            space *= -1
            # Horizontally align label to the right
            ha = 'right'
        # Use X value as label and format number
        label = '{:,.01f}'.format(x_value)
        # Create annotation\n",
        plt.annotate(
            label,                      # Use `label` as label
            (x_value, y_value),         # Place label at bar end
            xytext=(space, 0),          # Horizontally shift label by `space`
            textcoords='offset points', # Interpret `xytext` as offset in points
            va='center',                # Vertically center label
            ha=ha,                      # Horizontally align label differently for positive and negative values
            color = 'b')            # Change label color to white
    plt.show()
```

```python
#ф-ия обзора вещественных признаков
def hist_with_stat_lines(df, df_col, dtitle, dxlabel, dylabel, bins = 0, kde=False):
    df_col_description =  df_col.describe()
    mean = df_col_description[1]
    std = df_col_description[2]
    df_min = df_col_description[3]
    q1 = df_col_description[4]
    median = df_col_description[5]
    q3 = df_col_description[6]
    df_max = df_col_description[7]
    iqr = (q3-q1)+0.1 
    left_border = q1 - iqr*1.5
    # усы обрезаются минимальным и максимальным значением параметра сооветственно
    if left_border < df_min:
        left_border = df_min
    right_border = q3 + iqr*1.5
    if right_border > df_max:
        right_border = df_max
        
    if bins == 0:
        bins = right_border - left_border
        if bins>100:
            bins = 100
        elif bins<1:
            bins = bins*20
        elif bins>1 and bins<5:
            bins = 5
        bins = int(bins)
    else:
        bins = bins
        
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
    fig.suptitle(dtitle)   
    fig = sns.histplot(df, x=df_col, bins=bins, kde=kde, ax=axes[0]  ) # binwidth = bins/10 #kde=True  edgecolor='g' bins=bins  binwidth = bins/50
    axes[0].set_title('Гистограмма распределения')
    fig.set(xlim=(left_border, right_border)) 
    ax = sns.boxplot(x=df_col, data=df, ax=axes[1])
    ax = sns.stripplot(x=df_col, data=df, color="orange", jitter=0.2, size=2.5, ax=axes[1])
    axes[1].set_title('"Ящик с усами +"')
    #plt.title(dtitle)
    #plt.xlabel(dxlabel)
    axes[0].set_ylabel(dylabel)
    axes[0].set_xlabel(dxlabel)
    axes[1].set_xlabel(dxlabel)
    #axes[0].set.
    #plt.ylabel(dylabel)
    #fig.supxlabel(dxlabel)
    axes[0].axvline(x=mean,linewidth=1, color='r', linestyle='-' , label= 'mean_value')
    axes[0].axvline(x=median,linewidth=1, color='b', linestyle='-', label= 'median_value')
    axes[0].axvline(x=q1,linewidth=1, color='g', linestyle='--',label= 'Q1_value')
    axes[0].axvline(x=q3,linewidth=1, color='g', linestyle='--',label= 'Q3_value')
    #plt.legend(loc = 'best')
    axes[0].legend(loc = 'best')
    #fig.suptitle(dtitle, fontsize = 13) ;
    return print ('Статистические параметры колонки данных' ,dxlabel,'\n' ,'\n',  df_col_description,'\n')# df_col_description
```

```python
#ф-ия обрезки знчений лежащих за пределами q1-3iqr и q3+iqr
def replaceiqr (df_col):
    df_col_description =  df_col.describe()
    q1 = df_col_description[4]
    q3 = df_col_description[6]
    iqr = q3-q1
    #опреелим границы q1-3iqr и q3+iqr
    q1iqr = q1-iqr*3
    q3iqr = q3+iqr*3
    df_col = df_col.apply(lambda x: q1iqr if x < q1iqr else (q3iqr if x>q3iqr else x)) 
    return df_col
```

## Подготовка данных

### Разбиение на выборки

```python
#разбиение на выборки
features = df_new.drop(['Price', 'Price_log'], axis=1) 

target = df_new['Price'] 
target_log = df_new['Price_log'] 
  
# Разделение на 3 набора данных train, val и test 
X_train, X_test, y_train, y_test = train_test_split(features, target,
                                                    test_size=0.2, 
                                                    random_state=RANDOM_STATE)
X_train, X_val, y_train, y_val = train_test_split(X_train,
                                                  y_train,
                                                  test_size=0.2,
                                                  random_state=RANDOM_STATE)

#отдельное разбиение тестирования SGD c  лог-таргетом
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(features,
                                                    target_log,
                                                    test_size=0.2, 
                                                    random_state=RANDOM_STATE)
X_train_log, X_val_log, y_train_log, y_val_log = train_test_split(X_train_log,
                                                    y_train_log,
                                                    test_size=0.2, 
                                                    random_state=RANDOM_STATE)



#сделаем копии нетрансформированных выборок для catboost
X_train_non_trans = X_train.copy()
X_test_non_trans = X_test.copy()
X_val_non_trans = X_val.copy()


X_train.shape, X_test.shape, X_val.shape, X_train_log.shape
```

### Кодирование выборок для SGD

```python
#Типы данных для транформации
ohe_columns = ['VehicleType', 'Gearbox', 'FuelType', 'Model', 'Brand', 'Repaired']
num_columns = ['Power', 'Kilometer', 'RegistrationYear']
```

```python
encoder  = OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False ) #sparse=False   sparse_output=False 
# список всех категориальных признаков из тренировочной выборки
cat_col_names = ohe_columns
# обучим encoder
encoder.fit(X_train[cat_col_names])
# кодирование тренировочной выборки
X_train_ohe = encoder.transform(X_train[cat_col_names])
# обучим и преобразуем категориальные признаки из тренировочной выборки 
#X_train_ohe = encoder.fit_transform(X_train[cat_col_names]) 
# выполним преобразование категориальных признаков в валидационной выборке
X_val_ohe = encoder.transform(X_val[cat_col_names])
# выполним преобразование категориальных признаков в тестовой выборке
X_test_ohe = encoder.transform(X_test[cat_col_names])
# сохраним в переменной encoder_col_names названия новых столбцов 
encoder_col_names = encoder.get_feature_names_out()   #get_feature_names()
# создаём тренировочный и тестовый датафреймы из закодированных и отмасштабированных данных
X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_val_ohe = pd.DataFrame(X_val_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)
                               
# Инициализация scaler`a
scaler = StandardScaler()
num_col_names = num_columns
# обучим и преобразуем количественные признаки из тренировочной выборки 
X_train_scaled = scaler.fit_transform(X_train[num_col_names])
# выполним преобразование количественных признаков в тестовой выборке
X_val_scaled = scaler.transform(X_val[num_col_names])
# выполним преобразование количественных признаков в тестовой выборке
X_test_scaled = scaler.transform(X_test[num_col_names])
# создаём тренировочный и тестовый датафреймы из закодированных и отмасштабированных данных
X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_val_scaled = pd.DataFrame(X_val_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)

# объединяем все преобразованные данные в тренировочный и тестовый датафреймы

X_train_ohe = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_val_ohe  = pd.concat([X_val_ohe, X_val_scaled], axis=1)                            
X_test_ohe = pd.concat([X_test_ohe, X_test_scaled], axis=1)
                              
X_train_ohe.shape, X_val_ohe.shape, X_test_ohe.shape
```

### Кодирование выборок для моделей lightGBM и  RandomForest и CatBoost

```python
#Закодируем методом .cat.codes категориальные признаки для модели lightGBM и  RandomForest
X_train_cat = X_train_non_trans.copy()
X_val_cat = X_val_non_trans.copy()
X_test_cat = X_test_non_trans.copy()
for i in cat_list:
    X_train_cat[i] = X_train_cat[i].astype('category').cat.codes
    X_val_cat[i] = X_val_cat[i].astype('category').cat.codes
    X_test_cat[i] = X_test_cat[i].astype('category').cat.codes
X_train_cat.head()
```
